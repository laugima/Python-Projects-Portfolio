```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecciona solo columnas numéricas
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Calcula la correlación
correlation_matrix = df_numeric.corr()

# Crea el heatmap con la matriz de correlación
plt.rcParams['figure.figsize'] = (20,7)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)

plt.show()
```

```python
# Convertir la columna '2022 Population' a numérico y reemplazar errores por NaN
df["2022 Population"] = pd.to_numeric(df["2022 Population"], errors="coerce")

# Opcional: Verificar si otras columnas que necesitas son numéricas
# Si tienes más columnas que deben ser numéricas, convierte también esas columnas
# por ejemplo:
# df["OtraColumna"] = pd.to_numeric(df["OtraColumna"], errors="coerce")

# Eliminar filas con valores NaN en '2022 Population' para evitar errores en el cálculo
df = df.dropna(subset=["2022 Population"])

# Asegúrate de que solo tengas columnas numéricas en el DataFrame después del agrupamiento
# Aquí agrupamos y aplicamos 'mean' solo a las columnas numéricas
numeric_columns = df.select_dtypes(include=["number"]).columns
result = df.groupby('Continent')[numeric_columns].mean().sort_values(by="2022 Population", ascending=False)

print(result)
```
